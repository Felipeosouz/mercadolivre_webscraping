```python
import pandas as pd
import re

df = pd.read_csv("../data/data.csv", dtype={"preco": str})

df.head()
```

```python
df.info()
```

```python
df["preco"] = df["preco"].str.replace(".", "").astype(float)
df["km"] = df["km"].str.replace(" Km", "").str.replace(".", "").astype(int)
df.head()
```

```python
df["localização"].unique()
```

```python
def tratar_cidade(localizacao):
    if " - " in localizacao:
        return localizacao.split(" - ")[1]
    return localizacao
df["localização"] = df["localização"].apply(tratar_cidade)
df["localização"].unique()
```

```python
df["historico/vistoria"].unique()
```

```python
def processar_historico_vistoria(valor):
    historico = 0
    vistoria = 0
    if pd.notna(valor):
        if 'Histórico disponível' in valor:
            historico = 1
        if 'Vistoriado' in valor:
            vistoria = 1
    return pd.Series([historico, vistoria])

# Aplicar a função e criar as novas colunas
df[['historico', 'vistoria']] = df['historico/vistoria'].apply(processar_historico_vistoria)
df.drop("historico/vistoria", axis=1, inplace=True)
df.head(10)
```

```python
df["vendedor"] = df["vendedor"].fillna("Vendedor indepedente")
df.head()
```

```python
df["marca"] = df["nome"].str.split(" ").str[0]
df.head()
```

```python
def extrair_motor(nome):

    match = re.search(r'\b\d+\.\d+\b', nome)
    if match:
        return match.group()
    return "Não informado"

df['motor'] = df['nome'].apply(extrair_motor)
df.head(10)
```

```python
df.info()
```

```python
df.to_csv("../data/data_tratado.csv", index=False)
```
